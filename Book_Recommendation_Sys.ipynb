```python
# Loading the Books Rating Dataset
BookRatings = pd.read_csv("/Users/baranidharan/Desktop/Dataset/BX-Book-Ratings.csv", sep = ";",error_bad_lines=False,encoding='latin1')
```

```python
# There are many unwanted data's which doesn't support the cluster
BookRatings["User-ID"].value_counts()
```

```python
# Dropping the unwanted data's
x = BookRatings["User-ID"].value_counts() >200
```

```python
# Traversing the index of the user rated above 200

y = x[x].index
y
```

```python
# These are the users with rating above 200
High_Ratings = BookRatings[BookRatings["User-ID"].isin(y)] 
```

```python
High_Ratings.head()
```

```python
High_Ratings.shape
```

```python
# Merging the books id  (ISBN)
booksData.head()
```

```python
Books_With_Rating = High_Ratings.merge(booksData, on="ISBN")
Books_With_Rating
```

```python
Numbers_of_Ratings = Books_With_Rating.groupby("Book-Title")["Book-Rating"].count()
```

```python
# Books with the count of rating
Numbers_of_Ratings.reset_index()
```

```python
Numbers_of_Ratings.rename({'Book-Rating':'Number_of_Rating'},inplace=True)
```

```python
Numbers_of_Ratings.head()
```

```python
Final_Book_Rating = Books_With_Rating.merge(Numbers_of_Ratings,on="Book-Title")
```

```python
Final_Book_Rating
```

```python
Final_Book_Rating[Final_Book_Rating["Book-Rating_y"]> 50] 
```

```python
Final_DataFrame_BR = Final_Book_Rating[Final_Book_Rating["Book-Rating_y"]> 50] 
```

```python
Final_DataFrame_BR.shape
```

```python
Final_DataFrame_BR.drop_duplicates(["Book-Title","User-ID"])
```

```python
# By using the clustering method with Pivot table metrix
# Planning to group the similar users and provide recommendation based on the similar taste

BookPivot_Table = Final_DataFrame_BR.pivot_table(columns="User-ID",index="Book-Title",values = "Book-Rating_x")
```

```python
BookPivot_Table
```

```python
BookPivot_Table = BookPivot_Table.fillna(0)
```

```python
# As clusterring takes place, there are lot of 0's so, CSR Matrics is used

from scipy.sparse import csr_matrix

Sparse_Book = csr_matrix(BookPivot_Table)
Sparse_Book
```

```python
from sklearn.neighbors import NearestNeighbors

#Brute Force kNN algorithm calculates the squared distances from each query feature vector to each reference feature vector in the training data set. 
NN_Model = NearestNeighbors(algorithm="brute")

NN_Model.fit(Sparse_Book)
```

```python
distance, suggestion = NN_Model.kneighbors(BookPivot_Table.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
# Suggestions has the index number of the books
for i in range(len(suggestion)):
    print(BookPivot_Table.index[suggestion[i]])
```
